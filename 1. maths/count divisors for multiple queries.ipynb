```python
def count_multiple_divisors(m):
    
    least_prime=[-1]*(1000)
    least_prime[0]=0
    least_prime[1]=1
    for i in range(2,1000):
        if(least_prime[i]==-1):
            least_prime[i]=i
            for j in range(i*i,1000,i):
                if(least_prime[j]==-1):
                    least_prime[j]=i
    print(least_prime)
    
    import math
    for i in range(m):
        n=int(input())
        res=1
        for j in range(2,int(math.sqrt(n)+1)):
            if(n%j==0):
                count=0
                while(n%j==0):
                    count+=1
                    n//=j
                res=res*(count+1)
        if(n>1):
            res=res*2
        print(res)
```

```python
count_multiple_divisors(5)
```

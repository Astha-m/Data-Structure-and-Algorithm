```python


def prime_factors_multi(m):
    
    prime_least=[-1]*(1000)
    prime_least[0]=0
    prime_least[1]=1
    for i in range(2,1000):
        if(prime_least[i]==-1):
            prime_least[i]=i
            for j in range(i*i,1000,i):
                if(prime_least[j]==-1):
                    prime_least[j]=i
    print(prime_least)
    
    
    for i in range(m):
        val=int(input())
        while(prime_least[val]!=1):
            print(prime_least[val],end=' ')
            val=val//prime_least[val]
        print()
prime_factors_multi(1)
```

```python
class Node:
    def __init__(self,val):
        self.key=val
        self.next=None
        
```

```python
class SinglyLinkedList:
    def __init__(self):
        self.head=None
        
    def insert_at_end(self,val):
        temp=Node(val)
        if(self.head==None):
            self.head=temp
        else:
            curr=self.head
            while(curr.next!=None):
                curr=curr.next
            curr.next=temp
        
    def traverse(self):
        curr=self.head
        while(curr!=None):
            print(curr.key,end=' ')
            curr=curr.next
    
    def insert_at_beg(self,val):
        temp=Node(val)
        if(self.head==None):
            self.head=temp
        else:
            temp.next=self.head
            self.head=temp
        
    def insert_at_pos(self,val,pos):
        
        temp=Node(val)
        
        if(pos==1):
            if(self.head==None):
                self.head=temp
            else:
                temp.next=self.head
                self.head=temp
        else:
            curr=self.head
            i=1
            while(curr.next!=None and i<pos-1):
                curr=curr.next
                i+=1
            if(curr.next==None):
                curr.next=temp
            else:
                temp.next=curr.next
                curr.next=temp
    
    def delete_first(self):
        if(self.head==None):
            return
        else:
            temp=self.head
            self.head=temp.next
            temp=None
    
    def delete_last(self):
        if(self.head==None):
            return
        else:
            curr=self.head
            while(curr.next.next!=None):
                curr=curr.next
            curr.next=None
    
    def delete_at_pos(self,pos):
        if(self.head==None):
            return
        else:
            if(pos==1):
                temp=self.head
                self.head=self.head.next
                temp=None
            else:
                curr=self.head
                i=1
                while(curr.next.next!=None and i<pos-1):
                    curr=curr.next
                    i+=1
                temp=curr.next.next
                curr.next=None
                curr.next=temp
    
    def search(self,x):
        if(self.head==None):
            return -1
        else:
            i=1
            curr=self.head
            while(curr!=None):
                if(curr.key==x):
                    return i
                curr=curr.next
                i+=1
            return -1
        
    def middle_ele(self):
        if(self.head==None):
            return -1
        else:
            slow=self.head
            fast=self.head
            
            while(fast!=None and fast.next!=None):
                slow=slow.next
                fast=fast.next.next
            return slow.key
            
    def nth_from_the_end(self,k):
        if(self.head==None):
            return -1
        else:
            curr=self.head
            for i in range(k-1):
                curr=curr.next
            #print(curr.key)
            slow=self.head
            fast=curr
            while(fast.next!=None):
                slow=slow.next
                fast=fast.next
            return slow.key
            
    def reverse(self):
        curr=self.head
        prev=None
        while(curr!=None):
            nextt=curr.next
            curr.next=prev
            prev=curr
            curr=nextt
        
        curr=prev
        while(curr!=None):
            print(curr.key,end=' ')
            curr=curr.next
        print()
        self.head=prev
        
    
    def remove_duplicates_from_sorted_ll(self):
        if(self.head==None):
            return
        else:
            curr=self.head
            while(curr!=None):
                if(curr.next!=None and curr.next.key==curr.key):
                    temp=curr.next.next
                    curr.next.next=None
                    curr.next=temp
                curr=curr.next
    
    
        
        
    def detect_loop(self):
        if(self.head==None):
            return False
        
        slow=self.head
        fast=self.head
        
        while(fast!=None and fast.next!=None):
            slow=slow.next
            fast=fast.next.next
            if(slow==fast):
                return True
        return False
    
    def detect_remove_cycle(self):
        if(self.head==None):
            return False
        slow=self.head
        fast=self.head
        while(fast!=None and fast.next!=None):
            slow=slow.next
            fast=fast.next.next
            if(slow==fast):
                break
        
        if(slow==head):
            while(fast.next!=slow):
                fast=fast.next
                fast.next=None
                return
        elif(slow!=fast):
            return False

        slow=self.head
        while(slow.next!=fast.next):
            slow=slow.next
            fast=fast.next
        fast.next=None
        return True
    
    def delete_node_with_only_ptr_givn(self,ptr):
        if(ptr==None):
            return
        if(ptr.next==None):
            ptr=None
            return
        temp=ptr.next.next
        ptr.key=ptr.next.key
        ptr.next.next=None
        ptr.next=temp
        
    
    def segregate_even_odd(self):
        evenstart=None
        evenend=None
        oddstart=None
        oddend=None
        curr=self.head
        if(curr==None):
            return evenstart,oddstart
        
        while(curr!=None):
            if(curr.key%2==0):
                if(evenstart==None):
                    evenstart=curr
                    evenend=curr
                else:
                    evenend.next=curr
                    evenend=evenend.next
            else:
                if(oddstart==None):
                    oddstart=curr
                    oddend=curr
                else:
                    oddend.next=curr
                    oddend=oddend.next
            curr=curr.next
        if(evenstart==None or oddstart==None):
            return self.head
        evenend.next=oddstart
        oddend.next=None
        return evenstart
        
        
    def intersection(self,lst1,lst2):
        
        len1=0
        len2=0
        curr1=lst1
        curr2=lst2
        while(curr1!=None):
            len1+=1
            curr1=curr1.next
        
        while(curr2!=None):
            len2+=1
            curr2=curr2.next
        
        curr1=lst1
        curr2=lst2
        if(len1>len2):
            
            for i in range(len1-len2):
                curr1=curr1.next
        elif(len2>len1):
            
            for i in range(len2-len1):
                curr2=curr2.next
        
        while(curr1!=None and curr2!=None and curr1!=curr2):
            curr1=curr1.next
            curr2=curr2.next
        
        if(curr1!=curr2):
            return None
        else:
            return curr1
        
    
    def reverse_ll_in_grp_k(self,k):
        isFirst=True
        prevFirst=None
        curr=self.head
        while(curr!=None):
            first=curr
            prev=None
            count=0
            while(curr!=None and count<k):
                nextt=curr.next
                curr.next=prev
                prev=curr
                curr=nextt
                count+=1
            
            if(isFirst==True):
                self.head=prev
                isFirst=False
            else:
                prevFirst.next=prev
            prevFirst=first
        
    
    def merge_sorted_lls(self,lst1,lst2):
        if(lst2==None):
            return lst1
        if(lst1==None):
            return lst2
        
        a=lst1
        b=lst2
        newhead=Node(-1)
        prev=newhead
        while(a!=None and b!=None):
            if(a.key<=b.key):
                if(newhead==None):
                    newhead=a
                    prev=newhead
                    
                else:
                    prev.next=a
                    prev=a
                a=a.next
            else:
                if(newhead==None):
                    newhead=b
                    prev=newhead
                    
                else:
                    prev.next=b
                    prev=b
                b=b.next
        
        while(a!=None):
            prev.next=a
            a=a.next
        
        while(b!=None):
            prev.next=b
            b=b.next
        
        return newhead
    
    def reverseLLPalindrome(self,ptr):
        prev=None
        curr=ptr
        while(curr!=None):
            nextt=curr.next
            curr.next=prev
            prev=curr
            curr=nextt
        return prev
    
    def palindrome_check(self):
        if(self.head==None):
            return
        
        slow=self.head
        fast=self.head
        while(fast.next!=None and fast.next.next!=None):
            slow=slow.next
            fast=fast.next.next
        
        rp=reverseLLPalindrome(slow.next)
        
        slow=self.head
        while(rp!=None):
            if(slow.key!=rp.key):
                return False
            slow=slow.next
            rp=rp.next
        return True
    
    
sll=SinglyLinkedList()

sll.insert_at_end(10)
sll.insert_at_end(20)
sll.insert_at_end(30)
sll.insert_at_end(40)
sll.insert_at_end(50)
sll.insert_at_end(60)
sll.insert_at_end(70)
sll.insert_at_end(80)
sll.insert_at_end(90)
sll.insert_at_end(100)

sll.insert_at_beg(0)

sll.insert_at_pos(35,5)

sll.delete_first()

sll.delete_last()

sll.delete_at_pos(4)

print('search result position',sll.search(40))

print('middle val',sll.middle_ele())

print('nth from end',sll.nth_from_the_end(4))

sll.reverse()

sll.reverse()

sll.remove_duplicates_from_sorted_ll()

sll.reverse_ll_in_grp_k(5)

sll.traverse()

```

```python
'''
https://practice.geeksforgeeks.org/problems/close-to-perfection1525/1?page=2&difficulty[]=0&category[]=Dynamic%20Programming&sortBy=submissions

Given an array arr[] of N Numbers. A Perfect Piece
is defined as a subarray such that the difference 
between the minimum and the maximum value in that 
range is at most 1. Find the Maximal Length Perfect Piece.

 

Example 1:

Input:
N = 4
arr[] = {8, 8, 8, 8}
Output:
4
Explanation:
The longest subarray is [1, 4]
where maximum=8 and minimum = 8 and
difference is 0, which is less than 1.
Its length is 4.
Example 2:

Input:
N = 11
arr[] = {5, 4, 5, 5, 6, 7, 8, 8, 8, 7, 6}
Output:
5
Explanation:
The longest subarray is [6, 10]
where maximum=8 and minimum = 7 and
difference is 1. Its length is 5. 
 

Your Task:
You don't need to read input or print anything. 
Your task is to complete the function longestPerfectPiece() 
which takes an Integer N and an array arr[] 
of length N as input and returns the maximal 
length Perfect Piece.

 

Expected Time Complexity: O(N*logN)
Expected Auxiliary Space: O(N)

 

Constraints:
1 <= N <= 105
1 <= arr[i] <= 105


'''
```

```python
def longestPerfectPiece(self, arr, n):
        i=0
        j=0
        res=0
        minstack=[]
        maxstack=[]
        while(j<n):
            
            
            while(minstack and minstack[0]<i):
                minstack.pop(0)
            
            while(maxstack and maxstack[0]<i):
                maxstack.pop(0)
                
            
            while(minstack and arr[minstack[-1]]>arr[j]):
                minstack.pop()
            
            minstack.append(j)
            
            while(maxstack and arr[maxstack[-1]]<arr[j]):
                maxstack.pop()
                
            maxstack.append(j)
            
            if(minstack and maxstack and arr[maxstack[0]]-arr[minstack[0]]>1):
                i=min(maxstack[0],minstack[0])+1
            else:
                res=max(res,j-i+1)

            j+=1
        return res
```
